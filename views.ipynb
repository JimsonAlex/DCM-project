```python
import os
import shutil

# Paths
source_folder = 'DCM data/clean_images/Negative_Verified' 
destination_psax_folder = 'PLAX'  
destination_plax_folder = 'PSAX_PM'     


views_to_move = {
    'PSAX PM': destination_psax_folder,
    'PLAX': destination_plax_folder
}


if not os.path.exists(destination_psax_folder):
    os.makedirs(destination_psax_folder)
if not os.path.exists(destination_plax_folder):
    os.makedirs(destination_plax_folder)

for patient_id in os.listdir(source_folder):
    patient_path = os.path.join(source_folder, patient_id)
    
    if os.path.isdir(patient_path):
        
        
        for view, destination_folder in views_to_move.items():
            view_path = os.path.join(patient_path, view)
            
        
            if os.path.exists(view_path):
                
                
                for file_name in os.listdir(view_path):
                    file_path = os.path.join(view_path, file_name)
                    
                   
                    if os.path.isfile(file_path):
                       
                        new_file_name = f"{patient_id}_{file_name}"
                        shutil.move(file_path, os.path.join(destination_folder, new_file_name))
                
                # Optionally, remove the original view folder if it is empty after moving the files
                if len(os.listdir(view_path)) == 0:
                    os.rmdir(view_path)
                
                print(f"Moved {view} images for patient {patient_id} to {destination_folder}")
            else:
                print(f"View {view} not found for patient {patient_id}")

print("Process completed!")
```
